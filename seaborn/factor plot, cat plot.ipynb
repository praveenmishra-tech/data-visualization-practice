```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",kind="bar")
#scattar plot our factor plot ma difference
#kind="box" ka use karka koi bhe plot laga sakta hai and name dalna hoga
```
