```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#line sapace ek range aa jati hai bss our(1,ek ,10,our 20 data cheya)
var= np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var,vmin=0,vmax=12,cmap="PuOr",annot=True)
#cmap== color change kar sakt hai
#annot==matlab value jo data ma de gye hai uskoo apana oupt ma laha sakta hai
```
