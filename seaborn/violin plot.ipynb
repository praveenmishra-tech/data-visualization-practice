```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

a violin plot plays a similar role as a box and whisker plot. 
its shows the distrubution of quantitative data across several levels of one (or more) categorical variables suc that those distrubution can be compared

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,palette="Accent",hue="time")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,palette="Accent",hue="time",order=["Dinner","Lunch"],inner="stick")
plt.show()
#inner="stick"
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue='sex',split=True,scale="width")
plt.show()
```

single violin plot

```python
sns.violinplot(x="smoker",data=var)
plt.show()
```
