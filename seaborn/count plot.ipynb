what the difference between conplot and barolot

>countplot the count of the number of record by catogry
>barplot plots a vallue or materic for each category (by default, bar plot the mean of a variable by catogry)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var =sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data = var,hue="smoker",color="b",saturation=0.4)
#saturation sa thoda color halka ho jta hai
```

bar plot

```python
sns.barplot(x="sex",y="size",data=var)
```
