```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([33,6,22,66,44,33])
y = np.array([3,6,2,6,4,3])
plt.scatter(x,y,c="r",alpha=0.5)
plt.title("row data",fontsize=15)
t = plt.colorbar()
t.set_label("color bar")
plt.show()
#new thing alpha is how to dark a color in data and limit(0.1 to 0.10)
#color bar (mast cheez hai pura color aa jata hai)===>color map very important
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
#isko accha sa samjhna hai samjha nhi aya hai 
```
