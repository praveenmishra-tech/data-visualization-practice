With Pyplot, you can use the bar() function to draw bar graphs:

```python
import matplotlib.pyplot as plt
import numpy as np
x = [2,3,4,5,6,7,]
y = [7,8,5,7,5,4,]
plt.bar(x,y)
plt.show()
```

f you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
x = [2,3,4,5,6,7,]
y = [7,8,5,7,5,4,]
plt.barh(x,y)
plt.show()
```

the bar() takes the keyword argument width to set the width of the bars:

```python
x = [2,3,4,5,6,7,]
y = [7,8,5,7,5,4,]
colors  =   ["r","g","b","y"]
plt.bar(x,y,width=0.3,color=colors)
plt.show()
```

```python
# X-axis par jo names dikhne hain
import numpy as np
x = ["Python", "C", "C++", "Java"]

# First data
y = [85, 90, 80, 50]

# Second data
z = [9, 7, 8, 5]

# Har language ki position ban rahi hai
p = np.arange(len(x))

# Bar ki width
width = 0.3

# Dusre graph ko thoda right shift kar diya
p1 = [i + width for i in p]

# X-axis ka label
plt.xlabel("Language", fontsize=12)

# Y-axis ka label
plt.ylabel("Number")

# Graph ka title
plt.title("Praveen Mishra")

# Pehla Bar Graph
plt.bar(p, y, width, label="Marks")

# Dusra Bar Graph
plt.bar(p1, z, width, label="Experience")

# X-axis par language names set karna
plt.xticks(p + width/2, x)

# Legend show karega
plt.legend()#matlab marks experience ko represent kar rha hai

# Graph show
plt.show()
# ya samjh nhi aya hai
```
