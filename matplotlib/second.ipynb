```python
import matplotlib.pyplot as plt
import numpy as np
```

Line Color
You can use the keyword argument color or the shorter c to set the color of the line:

```python
xpoint= np.array([22,3,44,23,42,24])
plt.plot(xpoint,c="g")
plt.show()
```

Line Width:
You can use the keyword argument linewidth or the shorter lw to change the width of the line.

```python
x = np.array([22,90,55,7,44,3])
plt.plot(x,linewidth=3.5)#you can use lw in shortcut
plt.show()
```

```python
x = np.array([44,66,8,44,66,4])
y = np.array([44,62,54,56,32,3])
plt.plot(x)
plt.plot(y)
plt.show()
```

matplotlib in label

```python
x = np.array([44,1,111,40,563,99])
y = np.array([3,1,11,6,13,9])
plt.plot(x,y)
plt.xlabel("area")
plt.ylabel("total length")
plt.show()
```

```python
#With Pyplot, you can use the title() function to set a title for the plot.
x = np.array([22,44,55,66,77,88])
y = np.array([3,1,11,6,13,9])
plt.plot(x,y)
plt.title("size chart in graph")
plt.xlabel("area")
plt.ylabel("total length")
plt.show()
```

Position the Title
You can use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
x = np.array([22,44,55,66,77,88])
y = np.array([3,1,11,6,13,9])
plt.plot(x,y)
plt.title("size chart in graph",loc="right")#this
plt.xlabel("area")
plt.ylabel("total length")
plt.show()
```

Add Grid Lines to a Plot
With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()#this

plt.show()
```

You can use the axis parameter in the grid() function to specify which grid lines to display.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis="y")#this

plt.show()
```
