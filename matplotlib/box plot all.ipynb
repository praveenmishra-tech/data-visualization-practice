box plot ka upar our nicha yali line ma jo hota hai usko(viskey plot==>min  max  bata hai)

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,vert=False)#vertix sa denote karta hai
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,notch=True,)#notch=True,
plt.show()
```

```python
#outler dekhna hai and mean dekhna hai
x = [10,20,30,40,50,60,70]
plt.boxplot(x,showmeans=True,sym="g")
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
x1 = [22,44,55,33,99,67,35]
y = [x,x1]
plt.boxplot(y,label=["python",""])#vertix sa denote karta hai
plt.show()
```

[2]stack plot or area plot

```python
x= [1,2,3,4,5,]
area= [4,5,3,5,2]
area1= [4,5,3,4,1]
area2= [4,5,3,9,4]
area3= [4,3,1,7,4]
l = ["area","area1","area2","area3"]
plt.stackplot(x,area,area1,area2,area3,labels=l)
plt.title("area mesurment")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

[3]step plot

```python
x = [1,2,3,4,5,6,]
y = [1,2,3,4,5,6,]
plt.step(x,y)
plt.grid()
plt.title("postition")
plt.show()
```

fill_between plot

```python
x = [2,3,4,5,6,78,99]#2,3,4,5,6,78,99
y = [2,3,4,5,6,78,99]#2,3,4,5,6,78,99
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```
