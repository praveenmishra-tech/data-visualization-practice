```python
x= ["python","c","c++","java"]
y =[85,20,30,50]
z =[22,33,44,55]
plt.xlabel("language",fontsize=10)#label ma aap title add kar sakta hai(fontsize bhe add kar sakta ho)
plt.ylabel("number")#same isma bhe hota hai
plt.title("praveen mishra ")#heading or title add karna hai
plt.bar(x,y,width=0.4,label="praveen")#labela set karna hai graph to legnd use kkarna hai
plt.bar(x,z,width=0.4,label="praveen")
plt.legend()#legnd ma bhe parameter da sakta hai
plt.show()
```

```python
# X-axis par jo names dikhne hain
import numpy as np
x = ["Python", "C", "C++", "Java"]

# First data
y = [85, 90, 80, 50]

# Second data
z = [9, 7, 8, 5]

# Har language ki position ban rahi hai
p = np.arange(len(x))

# Bar ki width
width = 0.3

# Dusre graph ko thoda right shift kar diya
p1 = [i + width for i in p]

# X-axis ka label
plt.xlabel("Language", fontsize=12)

# Y-axis ka label
plt.ylabel("Number")

# Graph ka title
plt.title("Praveen Mishra")

# Pehla Bar Graph
plt.bar(p, y, width, label="Marks")

# Dusra Bar Graph
plt.bar(p1, z, width, label="Experience")

# X-axis par language names set karna
plt.xticks(p + width/2, x)

# Legend show karega
plt.legend()#matlab marks experience ko represent kar rha hai

# Graph show
plt.show()
```

```python
#Draw a line in a diagram from position (0,0) to position (6,250):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints,)
plt.show()
```

```python
#Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])


plt.plot(xpoints, ypoints,"*",color="r",label="praveen")
plt.legend()
plt.show()
```

The plot() function is used to draw points (markers) in a diagram.

marker="*"#you can use this keyword both of them are used plot and marker
marker size=  ms=20
marker edge and color= msc
Set the FACE color to red:mfc

```python
y = np.array ([82,33,46,55,90])
plt.plot(y,marker="*",ms=10,mec="r",mfc="y")
plt.show()
```

Format Strings (fmt)#shortcut bana ka liya phela (marker|line|color)

```python
y = np.array ([82,33,46,55,90])
plt.plot(y,"*:")
plt.show()
# '-'  	Solid line	
# ':'	Dotted line	
# '--'	Dashed line	
# '-.'	Dashed/dotted line
```
