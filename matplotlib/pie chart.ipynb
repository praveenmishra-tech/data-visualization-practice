With Pyplot, you can use the pie() function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array([35,67,34,77,])
plt.pie(x)
plt.show()
```

As you can see the pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]).

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:



label

```python
x= np.array([35,67,34,77,])
l =np.array(["apple","banan",'graphs','orange'])
plt.pie(x,labels=l)
plt.show()
```

Start Angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
x= np.array([35,67,34,77,])
l =np.array(["apple","banan",'graphs','orange'])
plt.pie(x,labels=l,startangle=90)
plt.show()
```

Explode

```python
x= np.array([35,67,34,77,])
l =np.array(["apple","banan",'graphs','orange'])
p= [0,0.1,0,0]
plt.pie(x,labels=l,startangle=90,explode=p,shadow=True)#adding shadow in explode value
plt.show()
```

```python
x= np.array([35,67,34,77,])
l =np.array(["apple","banan",'graphs','orange'])
p=[0.1,0.1,0.1,0.1]
plt.pie(x,labels=l,explode=p,shadow=True)
plt.legend(title="four fruit")
plt.show()
```

```python
x= np.array([35,67,34,77,])
l =np.array(["apple","banan",'graphs','orange'])
plt.pie(x,labels=l,startangle=90,autopct="%0.1f%%")
plt.show()
```

```python
x= np.array([35,67,34,77,])
x1= np.array([40,30,20,10])
l =np.array(["apple","banan",'graphs','orange'])
c= ["r","g","b","b"]
plt.pie(x,labels=l,colors=c,radius=1.5)
plt.pie(x1,colors="w")
plt.show()
```
