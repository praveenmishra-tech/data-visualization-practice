In Matplotlib, we use the hist() function to create histograms.

Standard Deviation batata hai ki data mean se kitna idhar-udhar phaila hua hai.

scale = Standard Deviation (Data kitna spread hoga) 
size = Kitni values generate karni hain

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(120,5,300)
#np.mean(x)
print(np.mean(x))
print(x)
```

```python
x = np.random.normal(120,5,300)
plt.hist(x)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
# X-axis = Values
# Y-axis = Frequency (Kitni baar value aayi)
```

```python
x = np.random.normal(120,5,300)
plt.hist(x,cumulative=-1,color="g")#cumulative==> frequency parameter ko reverse kar deta hai
plt.axvline(120,color="y",label="line")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.legend()
plt.show()
```
