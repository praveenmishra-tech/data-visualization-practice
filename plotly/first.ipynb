```python
import plotly.express as px
df = px.data.gapminder()
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
```

Setting size and color with column names

```python
var =px.data.iris()
fig =px.scatter(var,x="sepal_length",y="sepal_width",color="species",size="petal_length",hover_data="petal_width")
fig.show()
```

```python
#column length diya to bar color ban gya 
var=px.data.iris()
print(var)
fig = px.scatter(var,x="petal_length",y="petal_width",color="sepal_length")
fig.show()
```

```python
#the symbol argument can be mapped to a column as well. A wide variety of symbols are available.
var=px.data.iris()
print(var)
fig = px.scatter(var,x="sepal_length",y="sepal_width",color="species ")#,symbol="species"
fig.show()
```

Scatter plots and Categorical Axes

```python
var=px.data.medals_long()
fig =px.scatter(var,x="count",y="nation",color="medal",symbol="medal")
fig.update_traces(marker_size=10)
fig.show()
```

```python
#facetting
#Scatter plots support faceting.

import plotly.express as px
df = px.data.tips()
print(df)
fig = px.scatter(df, x="total_bill", y="tip", color="smoker", facet_col="sex", facet_row="time")
fig.show()
```

Linear Regression and Other Trendlines
>Scatter plots support linear and non-linear trendlines.

```python
df =px.data.tips()
fig =px.scatter(df,x="total_bill",y="tip",trendline="ols")
fig.show()
```
