```python
import plotly.express as px
import pandas as pd
```

Histograms with Plotly Expres
plotly express is the easy-to-use, high-level interface to Plotly,
which operates on a variety of types of data and produces easy-to-style figures.

```python
fig=px.data.tips()
var=px.histogram(fig,x="total_bill")
var.show()
```

```python
df = px.data.tips()
# Here we use a column with categorical data
fig = px.histogram(df, x="day")
fig.show()
```

Choosing the number of bins


```python
df = px.data.tips()
fig = px.histogram(df, x="total_bill", nbins=20)
fig.show()
```

Histograms on Date Data

```python
df = px.data.stocks()
fig = px.histogram(df, x="date")
fig.update_layout(bargap=0.2)
fig.show()
```

Histograms on Categorical Data

```python
df = px.data.tips()
fig = px.histogram(df, x="day", category_orders=dict(day=["Thur", "Fri", "Sat", "Sun"]))
fig.show()
```
