```python
import plotly.express as px
import pandas as pd
```

Area Chart, Line Chart jaisa hi hota hai, bas line ke neeche ka area color se fill ho jata hai.
>line_group ka kaam data ko alag-alag lines (ya area) me divide karna hai.

```python
df=px.data.gapminder()
print(df)
var=px.area(df,x="year",y="pop",color="country",line_group="continent")
var.show()
```

Ye plotly.graph_objects ka Filled Area Chart hai. 
Isme Scatter graph use hota hai, lekin line ke niche ka area bhi fill ho jata hai.

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[1, 2, 3, 4],
    y=[0, 2, 3, 5],
    fill='tozeroy'
))

fig.add_trace(go.Scatter(
    x=[1, 2, 3, 4],
    y=[3, 5, 1, 7],
    fill='tonexty'
))

fig.show()
```

facet_row     	Graph ko upar–neeche baant do
facet_col    	Graph ko left–right baant do
animation_frame	Graph ko chalne (animate) do
line_group  	Kis points ko ek line se jodna hai
fill	        Line ke niche color bhar do 

Stacked Area Chart kya hota hai?


Normal Area Chart → Ek line ke niche area fill hota hai.
Stacked Area Chart → Multiple categories ke filled areas ek doosre ke upar stack hote hain.
