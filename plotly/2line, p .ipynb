Line Plots with plotly.express

```python
import plotly.express as px
import pandas as pd
```

```python
#.query("country==Canada")#Ye sirf Canada ki rows ko select karta hai.
# #Iske baad df me sirf Canada ka data bachta hai
df = px.data.gapminder().query("country=='Canada'")
print(df)
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')
fig.show()
```

```python
df=px.data.gapminder().query("continent=='Oceania'")
print(df)
fig = px.line(df,x="year",y="lifeExp",color="country")
fig.show()

# df = px.data.gapminder().query("continent=='Oceania'")
# fig = px.line(df, x="year", y="lifeExp", color='country')
# fig.show()
```

Data Order in Line Charts

Plotly line chart => points ko usi order me connect karta hai jis order me data diya gaya hai.
Matlab line ulta-seedha zig-zag banne lagegi kyunki data sorted nahi tha.

Is problem ka solution

Graph banane se pehle data sort kar do.
df.sort("year")

```python
df= pd.DataFrame(dict(
    x=[1,4,2,6,5,3],
    y=[1,2,3,4,5,6]
))
fig=px.line(df,x="x",y="y",title="unsorted graph")
fig.show()

df =df.sort_values(by="x")
fig=px.line(df,x="x",y="y",title="sorted graph")
fig.show()


```

A connected scatter plot is a scatter plot in which the data points are connected by lines in a meaningful order (usually time), allowing us to visualize how the relationship between two continuous variables changes over time.

```python
df =px.data.gapminder().query("country in ['Canada', 'Botswana']")
fig=px.line(df,x="lifeExp", y="gdpPercap", color="country", text="year")
fig.update_traces(textposition="bottom left")
fig.show()
```

Line charts with markers

```python
df=px.data.gapminder().query("continent == 'Oceania'")
print(df)
var=px.line(df,x="year",y="lifeExp",color="country",markers=True)
var.show()
```

jab X-axis par Date ya Time hota hai aur Y-axis par koi numerical value hoti hai, tab us graph ko Time Series Line Chart bhi bolte hain.

```python
df=px.data.stocks()
print(df)
fig=px.line(df,x="date",y="GOOG")
fig.show()
```

```python
df=px.data.stocks(indexed=True)
print(df)
fig=px.line(df,facet_row="company",facet_row_spacing=0.1,height=200,width=200)#facet_row=>Har company ka alag mini graph banega.
                                                         #facet_row_spacingHar subplot ke beech bahut kam gap rakho.

# hide and lock down axes
fig.update_xaxes(visible=False ,fixedrange=True)
fig.update_yaxes(visible=False ,fixedrange=True)

# remove facet/subplot labels
fig.update_layout(annotations=[],overwrite=True)

# strip down the rest of the plot
fig.update_layout(showlegend=False,plot_bgcolor="white",margin=dict(t=10,l=10,b=10,r=10))

# disable the modebar for such a small plot
#displayModeBar=False==>koi option nhi dekha ga 
fig.show(config=dict(displayModeBar=False))
```
