```python
import plotly.express as px
import pandas as pd
```

```python
var=px.data.gapminder().query("country == 'Canada'")
print(var)
fig=px.bar(var,x="year",y="pop")
fig.show()
```

Bar charts with Long Format Data

```python
var=px.data.medals_long()
print(var)
fig=px.bar(var,x="nation",y="count",color="medal",title="this graph show medal")
fig.show()
```

Colored Bars

```python
var=px.data.gapminder().query("country == 'Canada'")
print(var)
fig=px.bar(var,x="year",y="pop",hover_data=['lifeExp','gdpPercap'],color="lifeExp",labels={'pop':'population of Canada'}, height=400)
fig.show()
```

Stacked vs Grouped Bars

```python
var=px.data.tips()
print(var)
fig=px.bar(var,x="sex",y="total_bill",color="time")
fig.show()
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.bar(df, x="sex", y="total_bill",
             color='smoker', barmode='group',
             height=400)
fig.show()
```

Aggregating into Single Colored Bars


```python
df = px.data.tips()
fig = px.histogram(df, x="sex", y="total_bill",
             color='smoker', barmode='group',
             height=400)
fig.show()
```

Bar Charts with Text

```python
df = px.data.medals_long()
print(df)

fig = px.bar(df, x="medal", y="count", color="nation", text_auto="True",text="nation")
fig.show()
```

```python
var=px.data.medals_long()
fig=px.bar(var,x="medal",y="count",color="nation",text="nation")
fig.show()
```

Facetted subplots

```python
df = px.data.tips()
fig = px.bar(df, x="sex", y="total_bill", color="smoker", barmode="group",
             facet_row="time", facet_col="day",
             category_orders={"day": ["Thur", "Fri", "Sat", "Sun"],
                              "time": ["Lunch", "Dinner"]})
fig.show()
```
